Step 2: Load the Data

```python
# Load the CSV file into a Pandas DataFrame
data = pd.read_csv(r"C:\Users\amree\Downloads\Datapreprocessing.csv")

# Display the first few rows of the DataFrame
print(data.head())
```

Step 3: Remove Duplicate Rows

```python
# Remove duplicate rows
data_no_duplicates = data.drop_duplicates()

# Display the DataFrame after removing duplicates
data_no_duplicates.head()
```

Step 4: Handle Missing Values

```python
data_no_duplicates.isnull()
```

```python
# Check for missing values
data_no_duplicates.isnull().sum()
```

```python
# Fill missing values with the mean of their respective columns
data_filled = data_no_duplicates.fillna(data_no_duplicates.mean(numeric_only=True))

# Display the DataFrame after filling missing values
data_filled.head()
```

```python
# Check for missing values after filling
data_filled.isnull().sum()
```

```python
data_filled.notnull()
```

```python
data_filled.isnull().sum().sum()
```

Step 5: Calculate Summary Statistics

```python
# Calculate summary statistics for the numerical columns
summary_statistics = data_filled.describe()

# Display the summary statistics
summary_statistics
```
